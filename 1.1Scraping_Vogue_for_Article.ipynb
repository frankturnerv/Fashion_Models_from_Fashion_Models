```python
#200 is good, anything else, lets investigate further

def verify_page(designer,season):
    url= 'https://www.vogue.com/fashion-shows/'+season+'/'+designer+'/slideshow/collection/print'
    status=0
    time.sleep(1)
    res=requests.get(url)
    status=res.status_code
    return status


    
```

```python
def create_directory_if_none_exists(destination,designer,season):
    if destination not in os.listdir():
        os.mkdir('./'+destination)
    if designer not in os.listdir('./'+destination):
        os.mkdir('./'+destination+'/'+designer)
    if season not in os.listdir('./'+destination+'/'+designer):
        os.mkdir('./'+destination+'/'+designer + '/'+season)
    
```

```python
def scrape_article_from_show(designer_page,season,destination):
    #load in the URL
    url= 'https://www.vogue.com/fashion-shows/'+season+'/'+designer_page
    res=requests.get(url)
    time.sleep(1)
    data={'day':'',
         'year':'',
         'location':'',
         'author':'',
         'season':'',
         'designer': '',
         'url':''}
    df=pd.DataFrame(columns=data.keys())
    if not res.ok:
        print('{} threw an error code of {}'.format(designer_page,res))
        print(url)
    else:    
        #creates a soup object   
        print(designer_page + ' ' + season)
        soup = BeautifulSoup(res.content,'lxml')
        place_date=soup.find('div',{"class":"details--date"}).find('span').text.split(',')
        location=place_date[0]
        day=place_date[1]
        year=place_date[2]
        
        author=soup.find('div',{"class":"details--author"}).text
        if author[0:3]=='by ':
            author=author[3:]
       
        season = soup.find('span',{'class':'article-content--category'}).text
        designer = soup.find('span',{'class':'article-content--title'}).text
        article = soup.find('div',{'class':'text--expandable'}).text
        
        data={'url':url,
             'day':day,
             'year':int(year),
             'location':location,
             'author':author,
             'season':season,
             'designer': designer,
             'article': article,
             }
        return data

    
```
