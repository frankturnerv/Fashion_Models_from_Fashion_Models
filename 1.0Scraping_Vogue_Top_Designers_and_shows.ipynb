```python
def progressBar(value, endvalue, bar_length=20):

        percent = float(value+1) / endvalue
        arrow = '-' * int(round(percent * bar_length)) + '>'
        spaces = ' ' * (bar_length - len(arrow))

        print("\rPercent: [{0}] {1}%".format(arrow + spaces, int(round(percent * 100))),end='\r'),
        
```

```python
#200 is good, anything else, lets investigate further

def verify_page(designer,season):
    url= 'https://www.vogue.com/fashion-shows/'+season+'/'+designer+'/slideshow/collection/print'
    status=0
    time.sleep(1)
    res=requests.get(url)
    status=res.status_code
    return status


    
```

```python
def create_directory_if_none_exists(destination,designer,season):
    if destination not in os.listdir():
        os.mkdir('./'+destination)
    if designer not in os.listdir('./'+destination):
        os.mkdir('./'+destination+'/'+designer)
    if season not in os.listdir('./'+destination+'/'+designer):
        os.mkdir('./'+destination+'/'+designer + '/'+season)
    
```

```python
def scrape_all_images_from_show(designer_page,season,destination):
    #load in the URL
    url= 'https://www.vogue.com/fashion-shows/'+season+'/'+designer_page+'/slideshow/collection/print'
    res=requests.get(url)
    time.sleep(1)
    data={'fashion_image':'',
         'designer':'',
         'season':'',
         'look': ''}
    df=pd.DataFrame(columns=data.keys())
    if not res.ok:
        print('{} threw an error code of {}'.format(designer_page,res))
        print(url)
    else:    
        #creates a soup object   
        print(designer_page + ' ' + season)
        soup = BeautifulSoup(res.content,'lxml')
        pages=soup.find_all('div',{"class":"slide--container page-break"})
        for page in range (0,len(pages)):
            #url of the image
            fashion_image=pages[page].find_all('img')[1]['src']
            designer=pages[page].find("div", {'class':'print-info--designer'}).text
            season=pages[page].find("div", {'class':'print-info--season'}).text
            look=pages[page].find("div", {'class':'print-info--look'}).text
            data={'fashion_image':fashion_image,
                 'designer':designer,
                 'season':season,
                 'look':look,
                 }
            create_directory_if_none_exists(destination, designer,season)
            #saves the image
            urllib.request.urlretrieve(fashion_image, "./{}/{}/{}/{}_{}_{}.jpg".format(destination,designer,season,designer,season,look))
            #saves the metadata in a dataframe
            df=pd.concat([df,pd.DataFrame(data,index=[page],columns=data.keys())])
            progressBar(page, len(pages), bar_length=20)
    df.to_csv("./{}/{}/{}/_image_contents.csv".format(destination,designer,season,))
    return df
    
```

```python
#make sure all pages are up 
for i,show in enumerate(shows):
    for j,designer in enumerate(shows[show]):
        status=verify_page(designer,show)
        progressBar(i, len(shows), bar_length=20)
```
